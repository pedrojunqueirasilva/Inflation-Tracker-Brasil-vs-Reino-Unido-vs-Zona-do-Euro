# Macro Cycle Comparison: Brazil, UK, and Euro Area

## Core Question
How have Brazil, the UK, and the Euro Area behaved across recent economic cycles, and what does that reveal about inflation pressure, monetary policy, and cost-of-living conditions?

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

project_root = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
panel = pd.read_csv(project_root / 'data' / 'processed' / 'macro_panel.csv')
panel.head()
```

```python
fig = px.line(panel, x='year', y='inflation_pct', color='country', template='plotly_white',
              title='Inflation Trend Comparison')
fig.show()
```

```python
fig = px.line(panel, x='year', y='real_interest_rate_pct', color='country', template='plotly_white',
              title='Real Interest Rate Trend Comparison')
fig.show()
```

## Interpretation Guide
- Compare post-2021 inflation persistence across countries.
- Assess whether real rates moved in line with inflation pressure.
- Compare unemployment and growth trade-offs during tightening phases.
- Use exchange-rate index to discuss imported inflation exposure.

## Limitations
- Annual harmonisation improves comparability but smooths short-term volatility.
- Real interest rate is broader than central-bank policy rate.
